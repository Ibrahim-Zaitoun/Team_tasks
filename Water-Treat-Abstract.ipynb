```python
var_grouped_value= data_treat.groupby('VAR')['Value'].mean()
var_grouped_value
```

```python
var_grouped_value.plot()
plt.title('Value grouped with VAR')
plt.xlabel('VAR')
plt.ylabel('Value')
plt.show()
```

```python
plt.scatter(var_grouped_value.index,var_grouped_value)
plt.title('Value grouped with VAR')
plt.xlabel('VAR')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Convert categorical variable to numerical using one-hot encoding
one_hot_vars = pd.get_dummies(var_grouped_value.index)

# Concatenate the one-hot encoded variables with the mean values
X = pd.concat([one_hot_vars, var_grouped_value], axis=1)

# Prepare the target variable
y = X['Value'].dropna()
X = X.drop('Value', axis=1).dropna()

X
```

```python
one_hot_vars
```

```python
y
```

```python
# Fit linear regression model
regressor = LinearRegression()
regressor.fit(X, y)

# Predict values using the linear model
y_pred = regressor.predict(X)

# Plot the scatter plot
plt.scatter(var_grouped_value.index, y, label='Data Points')

# Plot the regression line
plt.plot(var_grouped_value.index, y_pred, color='red', label='Regression Line')

plt.xlabel('VAR')
plt.ylabel('Mean Value')
plt.title('Scatter Plot with Linear Regression')
plt.legend()
plt.xticks(rotation= 45)
plt.show()
```

```python
data_treat.head()
```

```python
Country_grouped_year= data_treat.groupby(['Country','Year'])['Value'].mean()
Country_grouped_year
```

```python
# Convert the grouped data into a DataFrame
df_grouped = Country_grouped_year.reset_index()

# Create a bar plot using seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='Country', y='Value', hue='Year', data=df_grouped, palette='tab10')

plt.xlabel('Country')
plt.ylabel('Mean Value')
plt.title('Bar Plot of Grouped Data by Country and Year')
plt.legend(title='Year')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error


# Group the data by 'Country' and 'Year' and calculate the mean
Country_grouped_year = data_treat.groupby(['Country', 'Year'])['Value'].mean().reset_index()

# Prepare data for modeling
X = Country_grouped_year[['Country', 'Year']]
y = Country_grouped_year['Value']

# Convert categorical 'Country' feature to numerical using one-hot encoding
X_encoded = pd.get_dummies(X, columns=['Country'], drop_first=True)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=0)

# Create and train the Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=0)
rf_model.fit(X_train, y_train)

# Make predictions
y_pred = rf_model.predict(X_test)

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

# Visualize predicted vs. actual values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Value')
plt.ylabel('Predicted Value')
plt.title('Random Forest: Actual vs. Predicted')
plt.show()
```
